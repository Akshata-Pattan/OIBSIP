```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("online_retail_real_world.csv")

# Display first 5 rows
df.head()
```

```python
print("Shape:", df.shape)

print("\nDataset Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nSummary Statistics:")
display(df.describe(include='all'))
```

## Data Cleaning

In this step, the dataset is cleaned by handling missing values, removing duplicate records, converting data types, and standardizing text columns. These preprocessing steps improve data quality and prepare the dataset for further analysis.

```python
# Fill missing values

# Fill text columns with the most frequent value
df["ProductName"].fillna(df["ProductName"].mode()[0], inplace=True)
df["Brand"].fillna(df["Brand"].mode()[0], inplace=True)
df["Raw_Weight"].fillna(df["Raw_Weight"].mode()[0], inplace=True)

# Fill numeric column with median
df["UnitPrice"].fillna(df["UnitPrice"].median(), inplace=True)

# Verify
print(df.isnull().sum())
```

### Missing Values Handled

- Filled missing values in categorical columns using the most frequent value (mode).
- Filled missing values in numerical columns using the median.
- Verified that the dataset contains no missing values.

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Verify duplicates
print("Duplicate Rows:", df.duplicated().sum())
print("Shape after removing duplicates:", df.shape)
```

### Duplicate Records

Duplicate rows were checked and removed from the dataset. No duplicate records were found, indicating the dataset was already unique.

```python
# Convert OrderDate to datetime format
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

# Check data types
print(df.dtypes)
```

### Data Type Conversion

The `OrderDate` column was converted into datetime format to enable date-based analysis and improve consistency.

```python
# Remove extra spaces and standardize text columns

text_columns = ["ProductName", "Brand", "Country"]

for col in text_columns:
    df[col] = df[col].str.strip().str.title()

# Preview cleaned data
df[text_columns].head()
```

### Text Standardization

Text columns such as Product Name, Brand, and Country were cleaned by removing unnecessary spaces and converting values into a consistent title case format.

```python
# Save cleaned dataset

df.to_csv("Cleaned_Online_Retail_Dataset.csv", index=False)

print("✅ Cleaned dataset saved successfully!")
```

# Conclusion

The dataset was successfully cleaned and preprocessed by handling missing values, removing duplicate records, converting data types, and standardizing text fields. The cleaned dataset is now ready for Exploratory Data Analysis (EDA) and Machine Learning tasks.
